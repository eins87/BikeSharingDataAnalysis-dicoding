## Visualisasi data

```python
all_df.sample(5)
```

### Berapa banyak pengguna berdasarkan season/musim?
### Pada jam berapa rata-rata pengguna paling banyak per-hari dan per-minggu?

```python
all_df.head()
```

```python
weather_rentals = all_df.groupby('season_x')['cnt_y'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=weather_rentals, x='season_x', y='cnt_y')
plt.xlabel('Season/musim')
plt.ylabel('Rata-rata Pengguna')
plt.title('Rata-rat Bike Rentals berdasarkan season/musim')
plt.show()
```

```python
all_df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=all_df, x='hr', y='cnt_y')
plt.xlabel('jam dalam sehari')
plt.ylabel('Pengguna')
plt.title('Rata-rata Bike Rentals setiap jam')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=all_df, x='weekday_x', y='cnt_y')
plt.xlabel('Hari dalam seminggu')
plt.ylabel('pengguna')
plt.title('Rata-rata Bike Rentals setiap hari')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```
